## MCP Schema Reference

The `render_chart()` function accepts the **exact same JSON** as the `chart_render` MCP tool.

### Full ChartSpec Schema

```json
{
  "type": "line | bar | stacked_bar | scatter | bubble | area | histogram | pie | donut | surface_3d | candlestick | heatmap | waterfall | violin | boxplot",
  "title": "string (optional)",
  "subtitle": "string (optional)",
  "width": 595,
  "height": 300,
  "grid": true,
  "bins": 20,
  "series": [
    {
      "data": [1, 2, 3],
      "x": [0, 1, 2],
      "label": "Series name",
      "color": "#hex",
      "lineWidth": 1.5,
      "lineStyle": "solid | dashed | dotted | dashdot",
      "fillOpacity": 0.15,
      "markerSize": 4,
      "sizes": [5, 10, 15],
      "open": [], "high": [], "low": [], "close": []
    }
  ],
  "slices": [{"label": "A", "value": 10, "color": "#hex"}],
  "xLabels": ["A", "B", "C"],
  "hlines": [{"y": 0, "color": "#hex", "lineStyle": "dashed"}],
  "vlines": [{"x": 5, "color": "#hex", "lineStyle": "dashed"}],
  "annotations": [{"text": "Note", "x": 10, "y": 20, "color": "#hex"}],
  "legend": {"show": true, "position": "best"}
}
```

### MCP Tools

| Tool | Purpose |
|------|---------|
| `chart_render` | Render chart from ChartSpec JSON |
| `chart_resolve_type` | Recommend chart type from data description |
| `chart_get_styles` | Get theme, palettes, supported chart types |
