## 1. Line Charts

Use `type: "line"` for time series and trend data. Each series needs a `data` array. Optional: `color`, `lineWidth`, `lineStyle`, `fillOpacity`.

**Playground JSON** — paste into the editor at `localhost:3000`:
```json
{
  "type": "line",
  "title": "My Chart",
  "series": [{ "data": [0, 5, 12, 8, 18], "label": "Strategy", "color": "#d4d4d4" }],
  "xLabels": ["Jan", "Feb", "Mar", "Apr", "May"],
  "grid": true,
  "legend": { "show": true }
}
```

### 1a. Basic Line — Strategy vs Benchmark

````python
render_chart({
    "type": "line",
    "title": "Strategy vs Benchmark",
    "subtitle": "Cumulative returns (%)",
    "series": [
        {"data": [0, 5, 12, 8, 18, 25, 20, 32, 28, 40], "label": "Strategy", "color": "#d4d4d4"},
        {"data": [0, 3, 6, 5, 9, 12, 11, 15, 14, 18], "label": "Benchmark", "color": "#707070"}
    ],
    "xLabels": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"],
    "grid": True,
    "legend": {"show": True}
})
````

### 1d. Long Time Series — 60 Data Points

````python
import random
random.seed(7)

# Generate random walk data
strategy = [0]
benchmark = [0]
for _ in range(59):
    strategy.append(round(strategy[-1] + random.gauss(0.8, 3.0), 1))
    benchmark.append(round(benchmark[-1] + random.gauss(0.3, 1.2), 1))

render_chart({
    "type": "line",
    "title": "Dual Momentum Returns",
    "subtitle": "60 trading days — random walk simulation",
    "series": [
        {"data": strategy, "label": "Dual Momentum", "color": "#d4d4d4", "lineWidth": 1.8},
        {"data": benchmark, "label": "S&P 500", "color": "#707070"}
    ],
    "grid": True,
    "legend": {"show": True},
    "width": 700,
    "height": 280
})
````

## 7. Histograms

Use `type: "histogram"` with raw data in `series[0].data`. The engine auto-bins the data. Control bin count with `bins`.

**Playground JSON:**
```json
{
  "type": "histogram",
  "title": "Return Distribution",
  "series": [{ "data": [-3, -1, 0, 0.5, 1, 2, 3], "label": "Returns" }],
  "bins": 20,
  "grid": true
}
```

### 7a. Basic Histogram — Return Distribution

````python
import random
random.seed(42)
returns = [round(random.gauss(0.5, 2.0), 2) for _ in range(500)]

render_chart({
    "type": "histogram",
    "title": "Return Distribution",
    "subtitle": "Daily returns — 500 observations",
    "series": [{"data": returns, "label": "Returns"}],
    "bins": 25,
    "grid": True,
    "legend": {"show": True}
})
````

### 7b. Fine-Grained Histogram — 50 Bins

## 13. Matplotlib-like API

You can also use the `figure()` / `ax.subplot()` API directly — same engine, same output.

### 13a. Grouped Bar — Revenue vs Costs

````python
fig = figure(width=620, height=300)
ax = fig.subplot(1, 1, 1)
ax.bar(["Jan", "Feb", "Mar", "Apr", "May"], [12, 19, 8, 15, 22], label="Revenue", width=18)
ax.bar(["Jan", "Feb", "Mar", "Apr", "May"], [8, 14, 6, 11, 17], label="Costs", width=18)
ax.set_title("Revenue vs Costs")
ax.set_subtitle("Monthly breakdown (in $K)")
ax.grid(True)
ax.legend()
fig.show()
````

### 13b. Multi-Line — Strategy Comparison

````python
fig = figure(width=620, height=300)
ax = fig.subplot(1, 1, 1)
ax.plot([0, 5, 12, 8, 18, 25, 20, 32, 28, 40], color="#d4d4d4", label="Strategy", line_width=1.8)
ax.plot([0, 3, 6, 5, 9, 12, 11, 15, 14, 18], color="#707070", label="Benchmark")
ax.set_title("Cumulative Returns")
ax.set_subtitle("Strategy vs Benchmark (%)")
ax.set_xticks(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"])
ax.grid(True)
ax.legend()
fig.show()
````

### 13c. Scatter + Reference Lines

````python
fig = figure(width=620, height=300)
ax = fig.subplot(1, 1, 1)
ax.scatter([5, 8, 3, 12, 7, 15, 2, 10], [8, 12, 6, 15, 10, 18, 5, 14],
           color="#4ECDC4", label="Funds", size=6)
ax.axhline(10, color="#FFD93D", line_style="dashed")
ax.set_title("Risk vs Return")
ax.set_subtitle("Annualized metrics — 8 funds")
ax.grid(True)
ax.legend()
fig.show()
````

````python
fig = figure(width=400, height=300)
ax = fig.subplot(1, 1, 1)
ax.pie([35, 25, 20, 12, 8],
       labels=["Equities", "Bonds", "Real Estate", "Commodities", "Cash"],
       colors=["#4aaaba", "#d8b4fe", "#fbbf24", "#f9a8d4", "#6dd5c8"])
ax.set_title("Portfolio Allocation")
ax.legend()
fig.show()
````

### 13e. Histogram

````python
import random
random.seed(42)
data = [round(random.gauss(0.5, 2.0), 2) for _ in range(300)]

fig = figure(width=620, height=300)
ax = fig.subplot(1, 1, 1)
ax.hist(data, bins=25, label="Daily Returns", color="#C084FC")
ax.set_title("Return Distribution")
ax.set_subtitle("300 observations")
ax.axvline(0, color="#FF6B6B", line_style="dashed")
ax.grid(True)
ax.legend()
fig.show()
````

## ChartSpec JSON Schema Reference

The `render_chart()` function accepts the **exact same JSON** as the MCP `chart_render` tool and the Playground editor.

```json
{
  "type": "line | bar | stacked_bar | scatter | bubble | area | histogram | pie | donut | violin | boxplot | surface_3d | candlestick | waterfall | heatmap",
  "title": "string",
  "subtitle": "string",
  "width": 595,
  "height": 300,
  "grid": true,
  "bins": 20,

  "series": [
    {
      "data": [1, 2, 3],
      "x": [0, 1, 2],
      "label": "Series name",
      "color": "#hex",
      "lineWidth": 1.5,
      "lineStyle": "solid | dashed | dotted | dashdot",
      "fillOpacity": 0.15,
      "markerSize": 4,
      "sizes": [5, 10, 15]
    }
  ],

  "slices": [{"label": "A", "value": 10, "color": "#hex"}],
  "donutRatio": 0.55,

  "xLabels": ["A", "B", "C"],
  "hlines": [{"y": 0, "color": "#hex", "lineStyle": "dashed"}],
  "vlines": [{"x": 5, "color": "#hex", "lineStyle": "dashed"}],
  "annotations": [{"text": "Note", "x": 10, "y": 20, "color": "#hex"}],
  "legend": {"show": true}
}
```
